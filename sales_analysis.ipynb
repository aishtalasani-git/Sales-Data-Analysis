```python
import pandas as pd

df = pd.read_excel("data/Sample - Superstore.xls")

df.head()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

category_sales
```

```python
import matplotlib.pyplot as plt

category_sales.plot(kind="bar")

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.show()
```



**Observation:**
Technology generated the highest total sales, followed by Furniture. Office Supplies had the lowest sales.

**Business Insight:**
The company should continue investing in Technology products, as they contribute the most revenue.

```python
region_profit = df.groupby("Region")["Profit"].sum()

region_profit
```

```python
df.groupby("Customer Name")["Sales"].sum()
```

```python
segment_sales = df.groupby("Segment")["Sales"].sum()

segment_sales
```

```python
sub_category = df.groupby("Sub-Category")["Profit"].sum()

sub_category
```

```python
state_sales = df.groupby("State")["Sales"].sum()

state_sales
```

```python
top10_states = state_sales.sort_values(ascending=False).head(10)

top10_states
```

```python
import matplotlib.pyplot as plt

top10_states.plot(kind="bar")

plt.title("top10 states by sales")
plt.xlabel("state")
plt.ylabel("sales")

plt.show()
```

```python
df["Month"] = df["Order Date"].dt.to_period("M")
```

```python

monthly_sales = df.groupby("Month")["Sales"].sum()
monthly_sales


```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,8))

monthly_sales.plot(kind="line", marker="o")

plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.grid(True)

plt.show()
```


How do sales change over time?
Sales fluctuate throughout the years but show an overall increasing trend. The highest sales are recorded toward the end of 2017.

Business Insight
The increasing sales trend suggests business growth over time. The company may want to investigate why sales peak toward the end of the year and use those insights to improve performance in other months.

```python
product_sales = df.groupby("Product Name")["Sales"].sum()
product_sales
```

```python
top10_sales = product_sales.sort_values(ascending=False).head(10)

top10_sales
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,8))

top10_sales.plot(kind="bar")

plt.title("Top 10 Products by Sales")
plt.xlabel("Product Name")
plt.ylabel("Total Sales")

plt.xticks(rotation=90)

plt.grid(axis="y")

plt.show()
```


Which products generate the highest sales?
The top-selling product generated over $60,000 in sales, significantly higher than the remaining products.

Business Insight
The company should ensure these high-performing products remain in stock and consider promoting similar products to maximize revenue.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.scatter(df["Discount"], df["Profit"])

plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")

plt.grid(True)

plt.show()
```

```python
df[["Discount", "Profit"]].corr()
```

## Correlation Analysis

### Correlation Value
- Discount vs Profit = -0.2195

### Interpretation
There is a weak negative correlation between discount and profit. As the discount increases, profit tends to decrease, although other business factors also influence profitability.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,10))

# Chart 1
plt.subplot(2,2,1)
category_sales.plot(kind="bar")
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

# Chart 2
plt.subplot(2,2,2)
monthly_sales.plot(kind="line", marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)

# Chart 3
plt.subplot(2,2,3)
top10_states.plot(kind="bar")
plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.ylabel("Sales")
plt.xticks(rotation=90)

# Chart 4
plt.subplot(2,2,4)
plt.scatter(df["Discount"], df["Profit"])
plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.grid(True)

plt.tight_layout()
plt.show()
```

- West region earned the highest profit.
- Consumer segment contributed the highest revenue.
- California recorded the highest sales among all states.
- Sales showed an overall increasing trend over time.
- Higher discounts were generally associated with lower profits.
